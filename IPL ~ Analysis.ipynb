```python
import pandas as pd
```

```python
df = pd.read_csv('matches.csv')
```

```python
df.head()
```

```python
df['winner'].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.countplot(y=df['winner'])
plt.title('Number of Matches Won by Each Team')
plt.xlabel('Number of Wins')
plt.ylabel('Team')
plt.show()
```

```python
df['player_of_match'].value_counts().head(10)
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=df['player_of_match'].value_counts().head(10).values, y=df['player_of_match'].value_counts().head(10).index)
plt.title('Top 10 Players of the Match')
plt.xlabel('Number of Awards')
plt.ylabel('Player')
plt.show()
```

```python
toss_win_match_win = df['toss_winner'] == df['winner']
toss_win_match_win.value_counts()
```

```python
plt.figure(figsize=(7, 7))
plt.pie(toss_win_match_win.value_counts(), labels=['Yes', 'No'], autopct='%1.1f%%', startangle=90)
plt.title('Does the Toss-Winning Team Win the Match?')
plt.ylabel('') # Hides the 'None' label on the y-axis
plt.show()
```

```python
df['toss_decision'].value_counts()
```

```python
plt.figure(figsize=(7, 7))
plt.pie(df['toss_decision'].value_counts(), labels=df['toss_decision'].value_counts().index, autopct='%1.1f%%', startangle=90)
plt.title('Most Common Toss Decision')
plt.ylabel('')
plt.show()
```
